```python
import sqlite3
import pandas as pd

# Load datasets
athletes = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/athlete_events.csv")
regions = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/noc_regions.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep='\t')


# Connect to SQLite in-memory DB
conn = sqlite3.connect(":memory:")

# Write DataFrames to SQL tables
athletes.to_sql("athletes_table", conn, index=False, if_exists="replace")
regions.to_sql("regions_table", conn, index=False, if_exists="replace")
sales.to_sql("sales_table", conn, index=False, if_exists="replace")
```

```python
##  Question 1
```

```python
query_1 = """
SELECT r.region AS Country, COUNT(*) AS Total_medals
FROM athletes_table a
JOIN regions_table r ON a.NOC = r.NOC
WHERE a.Medal IS NOT NULL
GROUP BY r.region
ORDER BY total_medals DESC
LIMIT 5;
"""
pd.read_sql_query(query_1, conn)
```

```python
query_2 = """
SELECT ROUND(AVG(Age), 2) AS AVG_Gold_Medalist_Age
FROM athletes_table
WHERE Medal = 'Gold';
"""
pd.read_sql_query(query_2, conn)
```

```python
query_3 = """
SELECT Sport, COUNT(DISTINCT Event) AS Distinct_Events
FROM athletes_table
GROUP BY Sport
ORDER BY Distinct_Events DESC;
"""
pd.read_sql_query(query_3, conn)
```

```python
query = """
SELECT *
FROM athletes_table
WHERE NOC = 'USA';
"""

pd.read_sql_query(query, conn)
```

```python
query_5 = """
SELECT Year, COUNT(*) AS Medals_Awarded
FROM athletes_table
WHERE Medal IS NOT NULL
GROUP BY Year
ORDER BY Year DESC;
"""
pd.read_sql_query(query_5, conn)
```

```python
query_6 = """
SELECT *
FROM athletes_table
WHERE Height IS NULL
   OR Weight IS NULL;
"""

pd.read_sql_query(query_6, conn)
```

```python
#  Run the update query
update_query = """
UPDATE athletes_table
SET Height = (SELECT AVG(Height) 
FROM athletes_table)
WHERE Height IS NULL;
"""
conn.execute(update_query)
conn.commit()

# Verification check
verify_query = """
SELECT COUNT(*) AS missing_heights 
FROM athletes_table 
WHERE Height IS NULL;
"""
pd.read_sql_query(verify_query, conn)
```

```python
query_8 = """
SELECT Item_Name, 
       ROUND(
       SUM(quantity * CAST(
       REPLACE(Item_Price, '$', '') 
       AS REAL)), 2) AS Total_Sales
FROM sales_table
GROUP BY Item_Name
ORDER BY Total_Sales DESC;
"""
pd.read_sql_query(query_8, conn)
```

```python
query_9 = """
SELECT * FROM sales_table
ORDER BY CAST(REPLACE(item_price, '$', '') AS REAL) DESC
LIMIT 5;
"""
pd.read_sql_query(query_9, conn)
```

```python
query_10 = """
SELECT COUNT(DISTINCT order_id) AS Unique_Orders
FROM sales_table;
"""
pd.read_sql_query(query_10, conn)
```

```python
query_section_2 = """
WITH medalists AS (
    SELECT a.ID, a.Age, r.region AS country
    FROM athletes_table a
    JOIN regions_table r ON a.NOC = r.NOC
    WHERE a.ID IN (
        SELECT DISTINCT ID 
        FROM athletes_table 
        WHERE Medal IS NOT NULL
    )
)
SELECT 
    country,
    COUNT(DISTINCT ID) AS medal_winning_athletes,
    ROUND(AVG(Age), 2) AS average_age,
    CASE 
        WHEN AVG(Age) < 25 THEN 'High'
        WHEN AVG(Age) BETWEEN 25 AND 30 THEN 'Medium'
        ELSE 'Low'
    END AS performance
FROM medalists
GROUP BY country
ORDER BY medal_winning_athletes DESC;
"""

# Execute and display the data frame
pd.read_sql_query(query_section_2, conn)
```
